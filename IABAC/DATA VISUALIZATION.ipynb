# Domain Analysis

- **EmpNumber-** An Employee ID, also referred to as an Employee Number or Employee Code, is a unique number that has been assigned to each individual staff member within a company.

- **Age-** Age of employee.

- **Gender-** Gender of employee.

- **EducationBackground-** This feature represents the basic education such as, a high school diploma or a post-secondary degree of a employee.

- **MaritalStatus-** Marital status represents the fact of someone being married or not.
- **EmpDepartment-** The Department in which the employee works.

- **EmpJobRole-** Job role means the key responsibility of a job profile or job position of the employee.

- **BusinessTravelFrequency-** The Employee travel for company buisness purpose.

- **DistanceFromHome-** Distance between home of the employee to the company he/she works for.

- **EmpEducationLevel-** Employee Education level means the academic qualification, such as (diploma, degree, masters or PhD).

- **EmpEnvironmentSatisfaction-** Satisfied or happiness, employees are with elements like their jobs, their employee experience, and the organizations they are working for.

- **EmpHourlyRate-** Hourly Rate states the amount paid to an Employee for each hour worked.

- **EmpJobInvolvement-** Job involvement refers to a state of psychological identification with work—or the degree to which a job is central to a person's identity. From an organizational perspective, it has been regarded as the key to unlocking employee motivation and increasing productivity.

- **EmpJobLevel-** This feature indicates the Job levels, also known as job grades and classifications, set the responsibility level and expectations.

- **EmpJobSatisfaction-** Level of gratification Employees feel with their job.

- **NumCompaniesWorked-** This feature represents Employee worked in number of companies.

- **OverTime-** Overtime represents, Employee work overtime or not.

- **EmpLastSalaryHikePercent-** salary hike percent of employye in last year.

- **EmpRelationshipSatisfaction-** Healthy relationships may motivate employees and increase morale. When employees cast aside relationship issues, they can focus on work tasks more effectively.

- **TotalWorkExperienceInYears-** Total experiance of employee in years.

- **TrainingTimesLastYear-** Total training done by employee in last year.

- **EmpWorkLifeBalance-** This feature typically means the achievement by employees of equality between time spent working and personal life. A good work-life balance for employees can improve staff motivation, increase staff retention rates, reduce absence, attract new talent, and reduce employee stress.

- **ExperienceYearsAtThisCompany-** Total number of eaxperiance at current company.

- **ExperienceYearsInCurrentRole-** Total number of experiance in current job role.

- **YearsSinceLastPromotion-** Total number of year since last promotion of employee.

- **YearsWithCurrManager-** Employee total number of years with current manager.

- **Attrition-** Employee attrition is the naturally occurring, voluntary departure of employees from a company. Employee attrition involves leaving a job for Personal reasons.

- **PerformanceRating-** This feature tells the total rating of employees performance in the company.

# UNIVARIATE ANALYSIS:-
- Univariate analysis is the simplest kind of data analysis in the field of statistics. This could be either descriptive or inferential in nature as is the case in any data analysis in statistics. The key thing about the univariate analysis to remember is that there is only one data involved here. While the univariate analysis may be easy to analyze and also is not complex, at times it may end up giving some misleading information especially if there are more variables involved. In this case, you need to move to the bivariate and the multivariate analysis that will be capable of analyzing the data better.

###### VISUALIZATION OF CONTINUOUS FEATURES:-
**continuous features:**- It takes any measured value within a specific range. Continuous features can be visualized using different plot mentioned below.
- Histogram: A histogram is a graph used to show frequency distribution where data are grouped into continuous ranges and each range corresponds to a vertical bar. The horizontal axis represents the number range while the vertical axis represents the amount of data that is present in a range.

###### VISUALIZATION OF DISCRETE AND CATEGORICAL FEATURES:-

**Discrete and Categorical:**- Discrete features takes specific countable values. Categorical data is a type of data that is used to group information with similar characteristics.
-Discrete features can be visualized using different plots mentioned below.
- Countplot: A Countplot is used to represent the counts of observations in each categorical variables using bars.

### VISUALIZATION OF CATEGORICAL FEATURES:-
- USING SUBPLOT TO PLOTS MULTIPLE FEATURE

```python
count = d1[['Gender','EducationBackground','MaritalStatus','EmpDepartment','BusinessTravelFrequency']] # making lsit of next 12 feature

plt.figure(figsize=(15,20)) # defining canvas size
plotno = 1 # counter

for column in count: # iteration of columns / acessing the columns from count 
    if plotno<=13:    # set the limit
        plt.subplot(3,2,plotno) # # plotting 12 graphs (4-rows,3-columns) ,plotnumber is for count
        sns.countplot(x=count[column], palette = "magma") # Plotting count plots because the feature data type is discerte and categorical
        plt.xlabel(column,fontsize=20)  # assigning name to x-axis and font size is 20
    plotno+=1 # counter increment
plt.tight_layout()
plt.show() # used to hide the storage loction
```

**21)EducationBackground**
- The education background of employee in Life science and medical is maximum.

**22)MaritalStatus**
- The employee's with marital status married is maximum followed by single.
- The employee's with marital status divorced is minimum.

**23)EmpDepartment**
- The employee's with sales has the maximum counts.
