# Domain Analysis

**TxnID**: This stands for Transaction ID, and it is a unique identifier assigned to each transaction. It helps in tracking and distinguishing individual transactions in the system.

**Token**: In the context of financial transactions or authentication, a token is a piece of data that represents a user's authorization or access to perform specific actions. It is used as a security measure to validate and authorize transactions.

**TransactionType**: This attribute represents the type of transaction being performed. It could be a withdrawal, deposit, transfer, purchase, or any other type of financial transaction.

**Amount**: The amount attribute indicates the monetary value of the transaction, which could be positive (for deposits or transfers) or negative (for withdrawals or purchases).

**SenderID**: This refers to the identifier of the entity or user initiating the transaction. It uniquely identifies the sender or source of the transaction.

**PrevBalanceSender**: This represents the account balance of the sender before the transaction is processed. It provides a reference point to understand the impact of the transaction on the sender's account.

**CurrentBalanceSender**: This indicates the account balance of the sender after the transaction is processed. It helps track the updated balance after deducting the transaction amount and any associated fees.

**ReceiverID**: This attribute identifies the recipient or destination of the transaction. It uniquely identifies the recipient's account or user.

**PrevBalanceReceiver**: Similar to PrevBalanceSender, this represents the account balance of the receiver before the transaction is processed.

**CurrentBalanceReceiver**: This indicates the account balance of the receiver after the transaction is processed. It helps track the updated balance after adding the transaction amount.

**Time**: The time attribute records the timestamp when the transaction occurred. It helps in chronological ordering and tracking transaction history.

**Sender Location**: This attribute represents the geographical location or address of the sender at the time of the transaction. It could be useful for fraud detection or monitoring purposes.

**Receiver Location**: Similar to Sender Location, this attribute represents the geographical location or address of the receiver at the time of the transaction.

**Unnamed Column**: This is an unspecified column that could contain any other relevant information specific to the dataset or application. Its meaning and usage depend on the context of the data being analyzed.

In summary, the attributes provided represent essential information related to financial transactions, including transaction identifiers, types, amounts, sender and receiver details, account balances before and after transactions, timestamps, and location data. The IsSuspicious attribute adds a layer of security and fraud detection to the transactions. The unnamed column allows for additional data that might be present in the dataset but is not explicitly described in the list.

- Typically, any structured dataset includes multiple columns – a combination of numerical as well as categorical variables. A machine can only understand the numbers. It cannot understand the text. That’s primarily the reason we need to convert categorical columns to numerical columns so that a machine learning algorithm understands it. This process is called categorical encoding.

# Checking Correlation

- Correlation plots, also known as correlograms for more than two variables, help us to visualize the correlation between continuous variables.
- The range of correlation coefficients that are typically used to drop correlated features is between 0.7 and 0.9. However, this range can vary depending on the specific dataset and the machine learning model being used. In general, a correlation coefficient of 0.7 indicates that there is a strong linear relationship between two features, while a correlation coefficient of 0.9 indicates that there is a very strong linear relationship between two features.

- If two features are highly correlated, they are essentially providing the same information. This can lead to problems with machine learning models, such as:

- The model may not be able to learn the independent effect of each feature on the target variable.
- The model may be more sensitive to noise in the data.
- The model may be less generalizable to new data.
- By dropping highly correlated features, we can help to improve the performance of machine learning models.

```python
plt.figure(figsize=(20,20))
sns.heatmap(data.corr(),annot=True,cmap='rocket_r')
plt.show()
```

```python
# To get highly corelated feature 
pd.set_option('display.max_rows',None)
corrmat = data.corr()
corrmat = corrmat.abs().unstack()
corrmat = corrmat.sort_values(ascending=False)
corrmat = corrmat [corrmat >= 0.9] # Set threshold 0.9
corrmat = corrmat [corrmat < 1]
corrmat = pd.DataFrame(corrmat).reset_index()
corrmat.columns = ['feature1','feature2','corr']
corrmat
```

- There are no highly correlated features present in the dataset

```python
# #dropping highly correlated features
data.drop(['PrevBalanceSender','PrevBalanceReceiver'],axis=1,inplace=True)
```
